```python

words = macbeth.split()
```

```python

word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1 
```

```python
word_counts
```

```python
type(word_counts)
```

```python
df = pd.DataFrame.from_dict(word_counts, orient='index')
```

```python
df
```

```python
counts = df.sort_values(by=df.columns[0], ascending=False)
```
